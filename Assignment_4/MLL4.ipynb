```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('wine-quality.csv')
df.info()
```

```python
df.describe()
```

#Lets standerdize it

```python
scaler = StandardScaler()
df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
df.head()
```

```python
X = df.drop('quality', axis=1)
y = df['quality']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
model1 = LinearRegression()
for i in range(len(X_train.columns)):
  model1.fit(X_train.iloc[:, i].values.reshape(-1, 1), y_train)

model2 = LinearRegression()
model2.fit(X_train, y_train)
```

```python
print("Single Feature Linear Regression:")
for i in range(len(X_train.columns)):
  print(f"Feature: {X_train.columns[i]}")
  print(f"Parameter value: {model1.coef_[0]}")
  print(f"Training error: {model1.score(X_train.iloc[:, i].values.reshape(-1, 1), y_train)}")
  print(f"Test error: {model1.score(X_test.iloc[:, i].values.reshape(-1, 1), y_test)}")
  print()

print("Multiple Features Linear Regression:")
print(f"Parameter values: {model2.coef_}")
print(f"Training error: {model2.score(X_train, y_train)}")
print(f"Test error: {model2.score(X_test, y_test)}")
print(f"Model accuracy: {model2.score(X_test, y_test) * 100:.2f}%")
```
